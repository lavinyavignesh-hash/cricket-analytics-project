## Stage 2 – Data Cleaning and Pre-processing



maiden

```python
#Handle missing values
df_match["margin"]=df_match["margin"].fillna(df_match["margin"].mode()[0])
print(df_match.isnull().sum())
```

```python
#dropping the column
df_bowling=df_bowling.drop(columns=["maiden"])
print(df_bowling.columns)
```

```python
df_batting.columns
```

```python
df_player["bowlingStyle"]=df_player["bowlingStyle"].fillna(df_player["bowlingStyle"].mode()[0])
df_player["description"]=df_player["description"].fillna(df_player["description"].mode()[0])
df_player.drop(columns="image",inplace=True)
print(df_player.isnull().sum())
```

```python
#Treat outliers if required
sns.boxplot(x="runs",data=df_batting)
plt.title("Outliers of df_batting")
plt.show()
```

```python
sns.boxplot(x="runs",data=df_bowling)
plt.title("Outliers of df_bowling")
plt.show()
```

```python
#Check skewness and apply transformations
sns.histplot(x="runs",data=df_batting,kde=True)
plt.title("Checking skewness of df_batting")
plt.show()
```

```python
sns.histplot(x="runs",data=df_bowling,kde=True)
plt.title("Checking skewness of df_bowling")
plt.show()
```

```python
#Convert data types if needed
df_batting["SR"]=df_batting["SR"].replace("-",0)
df_batting["SR"]=df_batting["SR"].astype(float)

df_match["matchDate"]=pd.to_datetime(df_match["matchDate"])
```

## **Runs vs Balls Faced**

```python
sns.scatterplot(x='balls', y='runs', data=df_batting)
plt.title("Runs vs Balls Faced")
plt.xlabel("Balls Faced")
plt.ylabel("Runs Scored")
plt.show()
```

## What is it saying?

- As balls faced increases, runs scored also increases
Players facing more balls tend to score more runs

## Features used:

- Runs, Balls Faced

## Interpretation:

- There is a strong positive relationship between balls faced and runs scored
More opportunities lead to higher scoring

## What are we showing?

- Relationship between time spent batting and performance

## Insights:

- Data shows a positive correlation
- Some players score more runs with fewer balls (high efficiency)

## **Correlation Heatmap - Batting Performance**





```python
corr = df_batting[['runs', 'balls', '4s', '6s', 'SR']].corr()

sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap - Batting Performance")
plt.show()
```


## What is it saying?

- Runs have a strong positive correlation with balls (0.93)
- Runs are also highly related to 4s (0.87) and 6s (0.75)

## Features used:

- Runs, Balls, 4s, 6s, Strike Rate (SR)

## Interpretation:

- Players who face more balls tend to score more runs
Boundaries (4s & 6s) play a major role in scoring

## What are we showing?

- Relationship between batting factors and performance

## Insights:

- Strong correlation between runs and balls
- Strike rate has moderate correlation, indicating efficiency varies among players

## **Correlation Heatmap - Bowling Performance**

```python
corr = df_bowling[['overs', 'runs', 'wickets', 'economy', 'wides', 'noBalls']].corr()

sns.heatmap(corr, annot=True)

plt.title("Correlation Heatmap - Bowling Performance")

plt.show()
```

## What is it saying?
- Overs and runs have a moderate positive correlation (0.6)
- Economy has a negative correlation with wickets (-0.38)

## Features used:
- Overs, Runs, Wickets, Economy, Wides, NoBalls

## Interpretation:
- Bowlers who bowl more overs tend to concede more runs
- Better wicket-taking bowlers tend to have lower economy rates

## What are we showing?
- Relationship between bowling workload, performance, and efficiency

## Insights:
- Wickets and economy are negatively related (good performance)
- Extras (wides, noBalls) have weak correlation with wickets
